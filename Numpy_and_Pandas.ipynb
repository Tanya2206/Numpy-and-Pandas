TWO DIMENSIONAL ARRAY IN NumPy

```python
import numpy as np
sample_matrix=np.array(([50,20,1,23],[24,23,21,33],[56,76,24,7]))
sample_matrix
```

```python
print(sample_matrix[0][2])
sample_matrix[0][0]
#Counting of row and column starts from 0
#[row][column]
```

```python
sample_matrix[2,:]
#gives the complete 2nd row(all column elements included in that row)
```

```python
sample_matrix[2]
#columns not mention so same as above
```

```python
sample_matrix[:,(3,2)]
#rows not mention so all rows will be considered
#3,2 stands for column(3rd and 2nd column will be considered)
```

```python
sample_array=np.arange(1,20)
sample_array
```

```python
sample_array+sample_array
#element wise addition wiil be the output
```

```python
sample_array*3
#all the elements get multiplied with 3
#so mathematical operations are possible with the elements in array(can not be done like this in case of other data structures)
```

```python
sample_array/2
```

```python
np.exp(sample_array) 
#gives the exponential power of the elements
#(e^1,e^2,....)
```

```python
np.sqrt(sample_array)
#elementwise gives the square root values
```

```python
np.mean(sample_array)
#gives mean os all the elements in the array
```

```python
np.std(sample_array)
#satandard deviation
```

```python
np.var(sample_array)
#variance of the data
```
